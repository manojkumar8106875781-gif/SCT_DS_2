1. Import Libraries and Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel(r'C:\Users\vishn\OneDrive\Documents\train1.xlsx')
```

2. Univariate Analysis
A. Basic Info and Distribution

```python
print(df.describe(include='all'))  # Summary stats of all columns

# Age Distribution
plt.figure(figsize=(7,4))
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution')
plt.show()

# Survival Count
sns.countplot(x='Survived', data=df)
plt.title('Survival Counts')
plt.show()
```

3. Bivariate Analysis


A. Survival by Gender

```python
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Sex')
plt.show()
```

B. Survival by Passenger Class (Pclass)

```python
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Class')
plt.show()
```

C. Age vs. Survival

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age Distribution by Survival')
plt.show()
```

D. Fare vs. Survival

```python
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare Paid by Survivors vs Non-survivors')
plt.show()
```

E. Survival by Embarked Port

```python
sns.barplot(x='Embarked', y='Survived', data=df)
plt.title('Survival Rate by Embarkation Port')
plt.show()
```

4. Multivariate Analysis

A. Survival by Class and Gender

```python
sns.catplot(x='Pclass', hue='Sex', col='Survived',
            data=df, kind='count', height=4, aspect=.7)
plt.suptitle('Survival by Class and Gender', y=1.05)
plt.show()
```

5. Correlation Analysis

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='Blues')
plt.title('Correlation Between Numerical Variables')
plt.show()
```
